# 컬렉션 데이터 타입 : list, tuple, dict
- list [] : 순서가 있는 수정이 가능한 데이터 타입
- tuple () : 순서가 있는 수정이 불가능한 데이터 타입
- dist {} : 순서가 없고 키:값 으로 구성 되어 있는 데이터 타입

```python
# ls 
ls = [1, 2, 3]
# list 함수

ls.append(1)
print(ls)

ls.sort()
print(ls)

print(ls[::-1])

print(ls.pop())
print(ls)
```

```python
# 리스트 복사 
ls1 = [1, 2, 3]
ls2 = ls1 # 얕은 복사, 주소값 복사 
```

```python
ls1, ls2
```

```python
ls1[2] = 10
ls1, ls2
```

```python
ls3 = ls1.copy() # 깊은 복사
ls1, ls3
```

```python
ls1[2] = 30
```

```python
ls1, ls3
```

```python
tp1 = 1, 2, 3
tp2 = (4, 5, 6)
```

```python
type(tp1), type(tp2)
```

```python
tp1[2] = 3 
```

```python
a, b = 1, 2 
```

```python
# offset index 당연히 사용 가능 - 순서가 있으니깐 
tp1[1], tp2[::-1]

# 이것 조차도 튜플로 인식한다
```

```python
# list vs tuple 사이즈
import sys
ls = [1, 2, 3]
tp = 1, 2, 3

print(sys.getsizeof(ls), sys.getsizeof(tp))
```

# 연산자 
- 산술연산자 : +, -, , /, //, %, *
- 할당연산자 : 변수에 누적시켜서 연산 : +=, //=, **= ...
- 비교연산자 : >, <, ==, !=, <=, >= : 결과로 True, False
- 논리연산자 : True, False 를 연산 : or, and, not
- 멤버연산자 : 특정 데이터가 있는지 확인할때 사용 : not in, in

```python
# 논리 연산
True and False, True or False, not (True or False)
```

```python
# 멤버 연산
ls = ["jin", "andy"]
```

```python
"andy" in ls
```

```python
"me" not in ls
```
