# 파라미터와 아규먼트 
- 파라미터 : 함수를 선언할때 호출하는 부분에서 보내주는 데이터를 받는 변수
- 아규먼트 : 함수를 호출할때 함수에 보내주는 데이터 

```python
def plus(num1, num2=10, num3=20): #파라미터 : 디폴트 파라미터
    print(num1 + num2 - num3)
```

```python
plus(1, 2) # 아규먼트
```

```python
plus(3, num3=100) # 아규먼트 : 키워드 아규먼트
```

```python
print(1, 2, end="-") # 아규먼트와, 키워드 아규먼트를 보내준다!
print(1)
```

# 3. 리턴
- 함수의 결과를 저장하고 싶을때 사용
- return

```python
def plus(num1, num2):
    print(num1 + num2)
    return num1 + num2
```

```python
result = plus(1, 2)
print(result)
```

```python
# 리턴이 있는 것과 
# upper, lower, 
data1 = "python"
result = data1.upper()
print(result)
```

```python
# 리턴이 없는것
data2 = [3, 1, 2]
result = data2.sort()
print(result)
```

```python
data2
```

```python
# 함수 선언
def grade(point):
    if point >= 90:
        return "A"
    elif point >= 80:
        return  "B"
    else:
        return "C"
```

```python
point = 92
result = grade(point)
print(result)
```

```python
# 함수에서 return 코드가 실행되면 무조건 코드 실행이 종료 된다

def echo(msg):
    if msg == "quit":
        return
    print(msg)
```

```python
echo("python")
```

```python
echo("quit")
```

```python
# return은 튜플형으로 여러개 보낼수있어!
def calc(num1, num2):
    return num1 + num2, num1 - num2
data1, data2 = calc(1, 2)
data1, data2
```

### 4. `*args`, `**kwargs`
- 함수를 호출할때 아규먼트와 키워드 아규먼트 갯수를 특정 지을수 없을때 사용 

```python
def plus(*args, **kwargs):
    print(type(args), args)
    print(type(kwargs), kwargs)
    return sum(args) + sum(list(kwargs.values()))
```

```python
plus(1, 2, 3, 4, 5, num1=6, num2=7)
```

```python
# *args 는 tuple로 인식
# **kwargs 는 dict으로 인식 되네
```

```python
def func(num1, num2=10, num3=20):
    return num1 + num2 + num3

data = (1, 2, 3)
func(*data)
# func(data)
# func(1,2,3)
# func([1,2,3])
```

```python
data = {
    "num2": 100,
    "num3": 200,
}
func(1, **data) # func(1, num2=100, num3=200)
```
