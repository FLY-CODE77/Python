```python
# 지갑에 10000원 이상 있으면 택시를 탄다
# 2000원 이상 10000원 미만이면 버스를 탄다
# 안 그러면 걸어간다
```

```python
money = int(input())

if money >= 10000:
    print("택시를 타자")

elif money >= 2000:
    print("버스를 타자")

else:
    print("걸어가..")
```

# 삼항 연산자
- 간단한 if, else 구문을 한줄의 코드로 표현 가능한 방법
- (True) if (condition) else (False)

```python
data = 1 
if data:
    print("not zero")
else :
    print("zero")
```

```python
data = 0 
result = "not zero" if data else "zero"
result
```

# 반복문 
- 반복되는 코드를 실행
- while, for, break, continue
- list comprehention

```python
# while
data = 3 

while data :
    print(data)
    data -= 1
```

```python
# 무한 루프 
# break : 반복문 중단 시킬때 사용
result = 1 

while result:
    
    if result >= 10:
        break 
    result +=1

print(result)
```

# for 
- iterable한 값을 하나씩 꺼내서 variable 을 대입한후 코드를 iterable 변수의 값 만큼 실행

```python
# for: continue : 조건 부분으로 올라가서 코드가 실행
ls = [0, 1, 2, 3, 4]
for data in ls:
    if data %2 : #data 홀수이면
        continue
    
    print(data, end = " ")
```

# list comprehention
- 리스트 데이터를 만들어주는 방법
- for문과 유사하고, for문 보다 빠르다

```python
%time
ls = [0, 1, 2,  3]
result = []

for data in ls:
    result.append(data**2)

result
```

```python
%time
ls = [0, 1, 2,  3]
result = []

result = [data**2 for data in ls]
result
```

```python
ls = [1, 2, 3]
[func for func in dir(ls) if func[:2] != "__" and func[0] == "c"]
```

```python
# 리스트 컴프리헨션을써서 홀수와 짝수를 리스트로 출력해주는 코드
# 삼항연산 사용
ls = [0, 1, 2, 3]
# result = ["짝수", "홀수", "짝수", "홀수"]

result = [
    "홀수" if data % 2 else "짝수"
    for data in ls
]
result
```

```python
# 리스트 컴프리헨션 조건문
ls = range(10)
[data for data in ls if data%2]
```

```python
%%timeit
ls = []
for num in range(1, 100001):
    if num % 3 == 0:
        ls.append(num)
```

```python
%%timeit
ls = [num for num in range(1, 100001) if num %3 == 0]
```
